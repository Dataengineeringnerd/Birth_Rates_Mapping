```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from geopy.geocoders import Nominatim

# Read the birth rate data into a DataFrame
data = pd.read_csv('birth_rates_data.csv')

# Initialize a geocoder using Nominatim
geolocator = Nominatim(user_agent="my_geocoder")

# Create a Folium map centered around Lake County, Illinois
lake_county_map = folium.Map(location=[42.3684, -87.8272], zoom_start=10)

# Initialize a MarkerCluster for better marker management
marker_cluster = MarkerCluster().add_to(lake_county_map)

# Add markers for each ZIP code
for index, row in data.iterrows():
    # Geocode the ZIP code to obtain latitude and longitude
    location = geolocator.geocode(row['ZIP'])
    if location:
        lat, lon = location.latitude, location.longitude
        # Create a marker with custom icon and popup
        folium.Marker(location=[lat, lon],
                      popup=f"<b>ZIP: {row['ZIP']}</b><br>"
                            f"LBW: {row['LBW']}<br>"
                            f"Preterm: {row['Preterm']}<br>"
                            f"TeenBirth: {row['TeenBirth']}<br>"
                            f"Birth Rate: {row['Birth_Rate']}<br>"
                            f"F1stTriCare: {row['F1stTriCare']}<br>",
                      icon=folium.Icon(color='blue', icon='info-sign')).add_to(marker_cluster)

# Add a fullscreen button
folium.plugins.Fullscreen().add_to(lake_county_map)

# Add a layer control
folium.LayerControl().add_to(lake_county_map)

# Display the map
lake_county_map
```
